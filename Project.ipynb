```python
plt.style.use('classic')
plt.figure()
plt.subplot(2,1,1)
plt.plot(df["matchDuration"],"*")
print(plt.xlabel("Match Duration"))
plt.subplot(2,1,2)
plt.plot(df["walkDistance"],"--")
print(plt.xlabel("Walk Distance"))
plt.xticks(rotation=70);
plt.show()
```
